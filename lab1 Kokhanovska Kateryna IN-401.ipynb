```python
#з даних у рядку "top" можемо зробити висновок, що більшість студентів навчаються на 2 курсі, більшість студентів надають перевагу самостійному навчанню і не дуже активно беруть участь в позакласній діяльності. Найчастіше обирали мову програмування Python. Більшість студентів які брали участь в опитуванні ще не визначились стосовно свої подальший планів на майбутє, після закінчення навчання 
```

```python
#Фільтрація та отримання зрізів

df[df['Hours_of_study'] > 20 ] #студенти які навчаються більше ніж 20 годин на тиждень. обираємо стовпець який нас цікавить, в нашому випадку це "Hours_of_study", і ставимо умову фільтрації, у нас це мають бути всі рядки де значення цього стовпця більше 20, тобто ставимо "> 20"
```

```python
df.loc[df['Ave_score'] > 75, ['Student_ID', 'Age', 'Ave_score'] ] #виводимо лише ті рядки де середній бал більше 75, для рядків які підходять виводимо стовпці 'Student_ID', 'Age', 'Ave_score' 
```

```python
#тут ми застосували метод loc тому що нам потрібно було не тільки відфільтрувати дані з умовою а й вибрати окремі стовпці. І саме за допомогою цього методу це робити зручніше, бо ці дві задачі можна виконати разом
```

```python
df.loc[(df['Hours_of_study'] > 20 ) & (df['Satisfaction_level'] > 70), ['Hours_of_study', 'Preferred_study_method', 'Satisfaction_level']]
#виводимо інформацію про студентів які одночасно: витрачають на навчання більше ніж 20 годин на тиждень та їх задоволеність навчальною інфраструктурою більше ніж 70%. Для цих студентів виведимо лише стовпці 'Hours_of_study', 'Preferred_study_method', 'Satisfaction_level'.df.to_csv('DE-index.csv', index = 0)
```
